```python
import math
x_start = -31

def fun(x) -> float:
    return 169 * math.e ** (x / 13) - 13

def fun_first(x) -> float:
    return 13 * math.e ** (x / 13)


x = x_start
for k in range(0, 4):
    x_next = fun(x)
    x_next_first = fun_first(x)
    print(f"f(x) at {k} = {x_next}")
    print(f"f'(x) at {k} = {x_next_first}")

    x = x - (x_next / x_next_first)
```
